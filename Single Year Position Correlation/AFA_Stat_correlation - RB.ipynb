```python
import pandas as pd
data = pd.read_csv(r"C:\Users\JustinK\Downloads\AFA Data Cleanup 2022 - RB.csv")
df = pd.DataFrame(data, columns=['YDS', 'TD', 'ATT', 'AVG', 'FUM', 'LONG',
                                    'AGE',
                                    'SPEED', 'ACCELERATION', 'STRENGTH', 'AWARENESS',
                                    'AGILITY', 'CATCH', 'CARRY', 'JUMP', 'BREAK TACKLE', 'TACKLE',
                                    'PASS BLOCK', 'RUN BLOCK', 'TOUGHNESS', 'PLAYER EGO', 'WGT', 'HGTIN',
                                    'POWER','SPEED ROLE','BALANCED','POWER BACK','POSSESSION REC','SPEED BACK','NFL ICON','ELUSIVE BACK','FEATURE BACK'
                                    'TEAM LEADER','TEAM MENTOR','NONE'])
df = df.dropna(subset=['YDS'])
df = df[~(df['ATT'] <= 40)]
df['HGTIN'] = df['HGTIN'].astype(float)
correlation_matrix = df.corr()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(8,8))
rows = ['AGE','SPEED', 'ACCELERATION', 'STRENGTH', 'AWARENESS',
        'AGILITY', 'CATCH', 'CARRY', 'JUMP', 'BREAK TACKLE', 'TACKLE',
        'PASS BLOCK', 'RUN BLOCK', 'TOUGHNESS', 'PLAYER EGO', 'WGT', 'HGTIN',
        'POWER','SPEED ROLE','BALANCED','POWER BACK','POSSESSION REC','SPEED BACK','NFL ICON','ELUSIVE BACK','FEATURE BACK'
                                    'TEAM LEADER','TEAM MENTOR','NONE']
columns = ['YDS', 'TD', 'ATT', 'AVG', 'FUM', 'LONG']
sns.heatmap(correlation_matrix.loc[rows,columns], annot=True, cmap='PiYG', vmin=-.4,vmax=.7)
plt.title("2022 AFA RB Correlation Minimum 40 Attempts")
plt.tick_params(axis='both', which='major', labelsize=10, labelbottom = False, bottom = False, top = False, labeltop = True)
plt.show()
```
