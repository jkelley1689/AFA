```python
import pandas as pd
data = pd.read_csv(r"C:\Users\JustinK\Downloads\AFA Data Cleanup 2022 - RT.csv")
df = pd.DataFrame(data, columns=['PANCAKE', 'SACKS ALLOWED',
                                    'AGE',
                                    'SPEED', 'ACCELERATION', 'STRENGTH', 'AWARENESS',
                                    'AGILITY', 'CATCH', 'JUMP', 'BREAK TACKLE', 'TACKLE', 'PASS BLOCK', 'RUN BLOCK', 'STAMINA', 'TOUGHNESS', 'PLAYER EGO', 'WGT', 'HGTIN',
                                    'PASS BLOCKING','RUN BLOCKING','BALANCED','PASS BLOCKER','RUN BLOCKER','NFL ICON','ROAD BLOCKER','TEAM MENTOR',
                                    'TEAM LEADER','NONE'])
df = df.dropna(subset=['PANCAKE'])
df = df[~(df['PANCAKE'] <= 10)]
df['HGTIN'] = df['HGTIN'].astype(float)
correlation_matrix = df.corr()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(8,8))
rows = ['AGE','SPEED', 'ACCELERATION', 'STRENGTH', 'AWARENESS',
        'AGILITY', 'JUMP', 'BREAK TACKLE', 'TACKLE',
        'PASS BLOCK', 'RUN BLOCK', 'TOUGHNESS', 'PLAYER EGO', 'WGT', 'HGTIN',
        'PASS BLOCKING','RUN BLOCKING','BALANCED','PASS BLOCKER','RUN BLOCKER','NFL ICON','ROAD BLOCKER','TEAM MENTOR',
                                    'TEAM LEADER','NONE']
cols = ['PANCAKE', 'SACKS ALLOWED']
sns.heatmap(correlation_matrix.loc[rows,cols], annot=True, cmap='PiYG', vmin=-.4,vmax=.7)
plt.title("2022 AFA RT Correlation Minimum 10 Pancakes")
plt.tick_params(axis='both', which='major', labelsize=10, labelbottom = False, bottom = False, top = False, labeltop = True)
plt.show()
```
