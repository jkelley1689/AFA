```python
import pandas as pd
data = pd.read_csv(r"C:\Users\JustinK\Downloads\AFA Data Cleanup 2022 - WR (2).csv")
df = pd.DataFrame(data, columns=['REC', 'YDS', 'AVG', 'TD', 'YAC', 'DROPS', 'LONG',
                                    'AGE',
                                    'SPEED', 'ACCELERATION', 'STRENGTH', 'AWARENESS',
                                    'AGILITY', 'CATCH', 'CARRY', 'JUMP', 'BREAK TACKLE', 'TACKLE',
                                    'PASS BLOCK', 'RUN BLOCK', 'TOUGHNESS', 'WGT', 'HGTIN',
                                    'POSSESSION','SPEED_ROLE','BALANCED','POSSESSION REC','GO TO GUY',
                                    'TEAM DISTRACTION','TEAM LEADER','INJURY PRONE','UNDERACHIEVER',
                                    'DEEP THREAT','NONE'])
df = df.dropna(subset=['REC'])
df = df[~(df['REC'] <= 20)]
correlation_matrix = df.corr()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(8,8))
rows = ['AGE','SPEED', 'ACCELERATION', 'STRENGTH', 'AWARENESS',
        'AGILITY', 'CATCH', 'CARRY', 'JUMP', 'BREAK TACKLE', 'TACKLE',
        'PASS BLOCK', 'RUN BLOCK', 'TOUGHNESS', 'WGT', 'HGTIN',
        'POSSESSION','SPEED_ROLE','BALANCED','POSSESSION REC','GO TO GUY',
        'TEAM DISTRACTION','TEAM LEADER','INJURY PRONE','UNDERACHIEVER',
        'DEEP THREAT','NONE']
columns = ['REC', 'YDS', 'AVG', 'TD', 'YAC', 'DROPS', 'LONG']
sns.heatmap(correlation_matrix.loc[rows,columns], annot=True, cmap='PiYG', vmin=-.4,vmax=.7)
plt.title("2022 AFA Receiving WR Correlation Minimum 20 Catches")
plt.tick_params(axis='both', which='major', labelsize=10, labelbottom = False, bottom = False, top = False, labeltop = True)
plt.show()
```
