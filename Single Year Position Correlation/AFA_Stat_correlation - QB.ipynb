```python
import pandas as pd
data = pd.read_csv(r"C:\Users\JustinK\Downloads\AFA Data Cleanup 2022 - QB.csv")
df = pd.DataFrame(data, columns=['CMP', 'ATT', 'YDS', 'PCT', 'YPA', 'SACK', 'TD', 'INT', 'RATING',
                                    'AGE',
                                    'SPEED', 'ACCELERATION', 'STRENGTH', 'AWARENESS',
                                    'AGILITY', 'CATCH', 'JUMP', 'THROW POWER', 'THROW ACC', 'BREAK TACKLE', 'TACKLE', 'TOUGHNESS', 'PLAYER EGO', 'WGT', 'HGTIN'])
df = df.dropna(subset=['CMP'])
df['INT'] = df['INT']*(-1)
df = df[~(df['ATT'] <= 100)]
df['HGTIN'] = df['HGTIN'].astype(float)
correlation_matrix = df.corr()
df.to_clipboard()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(8,8))
rows = ['AGE','SPEED', 'ACCELERATION', 'STRENGTH', 'AWARENESS',
        'AGILITY', 'CATCH', 'JUMP', 'THROW POWER', 'THROW ACC', 'BREAK TACKLE', 'TACKLE', 'TOUGHNESS', 'PLAYER EGO', 'WGT', 'HGTIN']
cols = ['CMP', 'YDS', 'PCT', 'YPA', 'SACK', 'TD', 'INT', 'RATING']
sns.heatmap(correlation_matrix.loc[rows,cols], annot=True, cmap='PiYG', vmin=-.4,vmax=.7)
plt.title("2022 AFA QB Correlation Minimum 100 Attempts")
plt.tick_params(axis='both', which='major', labelsize=10, labelbottom = False, bottom = False, top = False, labeltop = True)
plt.show()
```
